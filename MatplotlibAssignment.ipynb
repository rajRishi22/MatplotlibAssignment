```python
# 1.Ans
# Matplotlib is an inbuilt python library that has functions for Data Visualization.Bar plot,scatter plots,line plot,Pie plot and Histogram plot.
```

```python
# 2.Ans
# A scatter plot is a 2D graph that takes two inputs x and y and plots a point in the graph.
import numpy as np
import matplotlib.pyplot as mp
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
mp.xlabel("X--Axis")
mp.ylabel("Y--Axis")
mp.title("Scatter Graph of the Given Data")
colors=np.random.rand(50)
sizes=800*np.random.rand(50)
mp.scatter(x,y,c=colors,s=sizes)
```

```python
# 3.Ans
# A Subplot Function is used to plot different graphs on the same screen without overlapping.
x=np.array([0,1,2,3,4,5])
y=np.array([0,100,200,300,400,500])
mp.subplot(2,2,1)
mp.plot(x,y)

x=np.array([0,1,2,3,4,5])
y=np.array([50,20,40,20,60,70])
mp.subplot(2,2,2)
mp.plot(x,y)

x=np.array([0,1,2,3,4,5])
y=np.array([0,10,20,30,40,50])
mp.subplot(2,2,3)
mp.plot(x,y)

x=np.array([0,1,2,3,4,5])
y=np.array([200,350,250,550,450,150])
mp.subplot(2,2,4)
mp.plot(x,y)
```

```python
# 4.Ans
# A bar plot is used to represent numerical data corresonding to non numerical data in form of Bars or Rectangular shaped lines .For Example, We have a name of a Smartphone Brand and Sales corresponding to it.
company=np.array(["Apple","Microsoft","Google","AMD"])
profit=np.array([3000,8000,1000,10000])
mp.bar(company,profit)
```

```python
mp.barh(company,profit)
```

```python
# 5.Ans
# A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum).
# It is used to represent distribution of data across a selected measure
box1=np.random.normal(100,10,200)
box2=np.random.normal(90,20,200)
mp.boxplot(box1)
```

```python
mp.boxplot(box2)
```
